# 패키지 (package)

```python
# import Mymodules.mod1  <-- 라고 하면..
# 경로에 해당하는 Mymodules 는 '패키지명' 이고, mod1 은 '모듈' 이다.

# 만약
# import aaa.bbb.dd.kkk  <-- 인 경우
# aaa.bbb.dd. 가 패키지명이고 kkk 가 모듈 이다.
# 즉 상대적인 경로로 aaa/bbb/dd  디렉토리 밑의 kkk.py 파이썬 파일을 불러오는 것이다
```

# as
긴 ~ 모듈명을 짧은 이름으로 명명하여 사용 가능

```python
import Mymodules.mod1 as my
```

```python
my.sum(20, 30)
```

# from 모듈 import (함수 / 변수 / 클래스 )

```python
# 모듈이름을 매번 붙이는게 번거롭다면
# 자주 사용하는 함수(변수, 함수, 클래스) 에 대해

#  다음과 같이 불러올수 있다
#  from 모듈이름 import 변수이름 혹은 함수이름 혹은 클래스이름
#  => 모듈이름을 붙이지 않고도 사용 가능.
```

```python
from Mymodules.mod1 import sum
```

```python
sum(55, 100)  # 모듈이름을 붙이지 않고도 사용 가능.
```

```python
# 여러개의 이름을 동시에 import 가능
from mod0 import myname, myFunc
```
