## 변수 값 바꾸기 (swap)

```python
a, b = 11, 33
print('a=', a, ', b=', b)
temp = a
a = b
b = temp
print('a=', a, ', b=', b)
```

```python
a, b = 100, 300
print('a=', a, ', b=', b)
a, b = b, a
print('a=', a, ', b=', b)
```
