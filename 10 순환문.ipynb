# while 순환문

```python
# while 순환문 구조,  조건식이 '참' 인 동안 수행문장을 반복수행

# while <조건식>:
#     <수행할 문장1>
#     <수행할 문장2>
#     <수행할 문장3>
#     ...
# else:
#      순환문 빠져나오기 전에 수행
```

```python
num = 0
while num < 10:
  print(num)
  num += 1

print('종료', num)
```

```python
# 순환문에서 중요한 것은
# 1. 몇번 순환을 하는가?
# 2. 순환하는 동안 변수값의 변화 범위는?
# 3. 순환문 종료후 변수값은?

# 위 순환문의 경우
# 1. 총 10번 순환을 했고
# 2. 순환하는 동안 num 변수값은 0 부터 9 까지 변화
# 3. 순환문 종료후 num 값은 10
```

```python
# 20 ~ 10 출력
num = 20
while num >= 10:
  print(num, end=' ')
  num -= 2

print('종료', num)
```

```python
# 구구단 2단을 출력해보자, while  사용
"""
2 x 1 = 2
2 x 2 = 4
2 x 3 = 6
2 x 4 = 8
2 x 5 = 10
2 x 6 = 12
2 x 7 = 14
2 x 8 = 16
2 x 9 = 18
"""
i = 1
while i <= 9:
  print(f'2 x {i} = {2 * i}')
  i += 1



```

# 중첩 순환문 (nested-loop)

```python
# 중첩 순환문 (nested loop)
# 조건문 안에 조건문 블럭이 들어갈수 있듯이
# 순환문 안에도 얼마든지 순환문이 포함될수 있다..
# 조건문과 순환문 의 어떠한 조합도 가능하다
```

```python
# 구구단 출력 : 2단 ~ 9단
# 중첩된 for 문

'''
2 x 1 = 2
2 x 2 = 4
...
2 x 9 = 18
3 x 1 = 3
...
4 x 1 = 4
…
...
9 x 9 = 81
'''
```

```python
dan = 2

while dan <= 9: # 2단 ~ 9단

  mul = 1
  while mul <= 9:   # x1 ~ x9
    print(dan, 'x', mul, '=', dan * mul)
    mul += 1

  dan += 1
```

# break

```python
# break
# 순환문 (while, for) 안 에서 순환문을 강제로 종료시키는 키워드

# break 는 감싸고 있는 가장 가까운 순환문을 종료합니다
```

```python
n = 1
while n <= 100:
  n += 1

  if n == 10:
    break
  print(n)



print('종료', n)
```

```python
# 연습
# 구구단 출력시
# 2단은 x 2 까지 출력
# 3단은 x 3 까지 출력
"""
2 단
2 x 1 = 2
2 x 2 = 4

3 단
3 x 1 = 3
3 x 2 = 6
3 x 3 = 9
...
"""
```

```python
dan = 2

while dan <= 9: # 2단 ~ 9단

  mul = 1
  while mul <= 9:   # x1 ~ x9
    print(dan, 'x', mul, '=', dan * mul)
    if dan == mul: break
    mul += 1

  dan += 1
```

# continue

```python
# continue
# 순환문 처음으로 돌아가기
# 순환문은 종료하지 않되, 특정 조건만 skip 하는 경우 사용
```

```python
num = 1
while num <= 10:
  num += 1
  if num == 6: break
  print(num)
```

```python
num = 1
while num <= 10:
  num += 1
  if num == 6: continue
  print(num)
```

```python
# 정수를 계속해서 입력받다가, 0 이 입력되면
# 그때까지의 정수의 '합' 과 '평균' 을 출력

"""
[입력예]
10
20
30
0

[출력예]
합: 60
평균: 20.0

"""
total = 0 # 합
cnt = 0 # 개수

while True:  # 무한루프
  n = int(input())
  if n == 0: break   # 0 입력하면 무한루프 종료
  total += n  # 합계 누적 합산
  cnt += 1   # 개수 누적

print('합:', total)
print('평균:%.1f' % (total / cnt))
```

## List Comprehension

```python
# List Comprehension 구문
# [표현식 for 항목 in 반복가능객체 (if 조건)]
```

```python
# 주어진 리스트의 원소 데이터 들을 * 3 하여 새로운 리스트 작성하기
# ex) [1, 2, 3, 4] ==> [3, 6, 9, 12]

# 일단 지금까지 배운 방법으로 한다면

a = [1, 2, 3, 4]
result = []
for num in a:
  result.append(num * 3)

print(result)
```

# 도전과제

```python
# 1부터 100 까지의 숫자 출력.  한줄에 10개씩 출력
# 숫자 중에 3, 6, 9 가 있으면 숫자대신 '*' 출력
# 단! 문자열 함수 쓰지 말기

"""
    1    2    *    4    5    *    7    8    *   10
   11   12    *   14   15    *   17   18    *   20
   21   22    *   24   25    *   27   28    *    *
    *    *    *    *    *    *    *    *    *   40
   41   42    *   44   45    *   47   48    *   50
   51   52    *   54   55    *   57   58    *    *
    *    *    *    *    *    *    *    *    *   70
   71   72    *   74   75    *   77   78    *   80
   81   82    *   84   85    *   87   88    *    *
    *    *    *    *    *    *    *    *    *  100
"""
None
```

```python
n = 100
i = 1
while i <= n:  # 1 ~ 100까지
  digit1 = i % 10  #  1의 자리
  digit2 = i // 10 # 10의 자리

  # 3, 6, 9 여부에 따른 출력
  if digit1 == 3 or digit1 == 6 or digit1 == 9 \
    or digit2 == 3 or digit2 == 6 or digit2 == 9:
    print(f'{"*":>5}', end='')
  else:
    print(f'{i:5}', end='')

  i % 10 or print()  # 10개 출력후 줄바꿈.
  i += 1
```
