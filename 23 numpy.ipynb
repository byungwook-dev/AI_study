# shape (차원) 변경하기
**차원변환**은 데이터분석 과 인공지능에서 매우 빈번하게 발생되는 작업이니만큼 자유자재로 변환할수 있어야 하고, 머리속으로 내가 변환하는 데이터의 구조가 그려지도록 익숙해져야 합니다

사용 예) 이미지 데이터 벡터화 - 이미지는 기본적으로 2차원 혹은 3차원(RGB)이나 트레이닝을 위해 1차원으로 변경하여 사용 됨

## ravel(), np.ravel()
  - [ˈrævl]
  - 다차원배열을 1차원으로 변경 (흔히 **'펼친다'**라고 말함)
  - 'order' 파라미터
    - 'C' - row 우선 변경 ,   C style
    - 'F - column 우선 변경,   Fortran style
  - ravel() 은 np 안에 일반 함수로도 있고, 혹은 ndarray 의 멤버함수로서도 존재

```python
x
```

```python
x.ravel()
```

```python
# np.ravel?
```

```python
np.ravel(x, order='C')   # row 방향으로 펼쳐짐.  (axis 0 부터 펼쳐짐)
```

```python
np.ravel(x, order='F')   # column 방향으로 펼쳐짐 (axis -1 부터 펼쳐짐)
```

## reshape 함수
 - array의 shape을 다른 차원으로 변경
 - 주의할점은 reshape한 후의 결과의 전체 원소 개수와 이전 개수가 같아야 가능
 - 사용 예) 이미지 데이터 벡터화 - 이미지는 기본적으로 2차원 혹은 3차원(RGB)이나 트레이닝을 위해 1차원으로 변경하여 사용 됨
 - **-1** 값을 주어 나머지 차원을 유추하게 할수 있다

```python
x2 = np.arange(36)
x2
```

```python
x2.reshape(6, 6)
```

```python
x2.reshape(6, -1)  # -1 의 차원 크기가 유추되어 reshape 된다 => (6,6)
```

```python
x2.reshape(6, -1, 2) # => (6, 3, 2)
```

## 차원 자동 제거 : squeeze, np.squeeze()

- 차원 중 사이즈가 1인 것을 찾아 스칼라값으로 바꿔 해당 '차원을 제거'한다.

# axis (차원축)

```python
"""
numpy 의 많은 함수들은 axis=  파라미터를 갖고 있습니다.
axis (축) 은 각 '차원' 을 의미합니다.
1차원 array 라면  각 차원에 대한 axis 값은 0  <- 1개 입니다.
2차원 array 라면  각 차원에 대한 axis 값은 0, 1  <- 2개 입니다.
3차원 array 라면  각 차원에 대한 axis 값은 0, 1, 2  <- 3개 입니다.

axis= 값을 명시하면 그 함수의 연산 은 해당 axis(축) 에 '따라서' 연산을 수행함

과연 '따라서' 의 의미가 무엇인지 실습을 통해 이해해봅시다
"""
None
```

```python
np.sum(x1)
```

```python
np.sum(x1, axis=None)
```

```python
x2 = x1.reshape(3, 5)
x2
```

```python
x2.shape

# (행, 열)
# (axis 0, axis 1)
```

```python
np.sum(x2)  # axis=None
```

```python
x2
```

```python
np.sum(x2, axis=0)  # (3, 5) => (5,)

# 15 = 0 + 5 + 10
#      [0,0] + [1, 0] + [2, 0]
```

```python
np.sum(x2, axis=1)  # (3, 5) => (3,)
```

```python
np.sum(x2, axis=-1)
```

 - [Rule 공식문서](https://docs.scipy.org/doc/numpy/user/basics.broadcasting.html#general-broadcasting-rules)
 - 뒷 차원에서 부터 비교하여 **①Shape이 같거나**, **②차원 중 값이 1이 존재**하면 broadcasting 가능
 - 결과 차원은 둘중 큰 size 의 차원으로 **확장** 된다



![브로드캐스팅 예](https://www.tutorialspoint.com/numpy/images/array.jpg)

# Boolean Indexing

```python
# 'Boolean indexing ' !
# array 인덱싱 시, bool 리스트(aka. mask)를 전달하여 True인 경우만 필터링

# for 사용하지 않고도 array 에서 '조건'에 맞는 데이터만 추출 하는 기능
```

```python
x = np.random.randint(1, 100, size=10)
x
```

```python
x % 2
```

```python
even_mask = x % 2 == 0
even_mask
```

```python
# 위와 같이 bool 값으로 이루어진 array 를 Mask 라고도 한다
# 위 결과, mask 를 변수에 담아 보겠습니다

# boolean mask 로 인덱싱!
x[even_mask]
```

```python
# list 나 tuple 같은 iterable 도 bool mask 로 사용 가능.

x[[ True,  True,  True,  True, False, False, False, False, False, True]]
```

```python
x[x % 2 == 0]
```

```python
x[x > 30]
```

## 다중조건 사용하기
 - 파이썬 논리 연산자인 and, or, not 키워드는 사용 불가
 - & ← AND
 - | ← OR

```python
# 짝수이면서 30보다 작은 숫자들

# x[(x % 2 == 0) and (x < 30)]

x[(x % 2 == 0) & (x < 30)]
```

## 도전: 서울시 5월 일자별 평균기온

```python
temp = np.array(
        [23.9, 24.4, 24.1, 25.4, 27.6, 29.7,
         26.7, 25.1, 25.0, 22.7, 21.9, 23.6,
         24.9, 25.9, 23.8, 24.7, 25.6, 26.9,
         28.6, 28.0, 25.1, 26.7, 28.1, 26.5,
         26.3, 25.9, 28.4, 26.1, 27.5, 28.1, 25.8])

"""
평균기온이 25도를 넘는 날수는 총 21일
평균기온이 25를 넘는 날의 평균 기온은 약 26.8도
"""
None
```

```python
days = np.sum(temp>25)
print(days)
avr = np.average(temp[temp>25])
print(avr)
```

```python
print(np.sum(temp > 25))


temp_avg = np.sum(temp[temp>25])/np.sum(temp > 25)
print(temp_avg)
```

```python
item = (temp > 25).astype(int).sum()
hap = temp[(temp > 25)].sum()
avg = hap / item
print(f"%.2f" % avg)
```

```python
len(temp[temp > 25]), sum(temp[temp > 25])/len(temp[temp > 25])
```

```python
temp[temp>25].sum()/np.sum(temp>25)
```

```python
day = (temp > 25).astype(int).sum()
print(day)
(temp[temp > 25].sum()) / day
```

```python
a = (temp>25.0).sum()
print("평균기온이 25도를 넘는 날수는 총",a,"일")
b = np.average(temp[a])
print("평균기온이 25를 넘는 날의 평균 기온은 약",b,"도")
```
