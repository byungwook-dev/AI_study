# Assignment Unpacking
tuple 은 iterable 객체다

대입연산자와 iterable 을 사용하여 여러 개의 변수에 대입 가능  (이를 Assignment unpacking 이라 함)

```python
# iterable 객체
# str, list, tuple, set, dict ... <- iterable 객체다
```

```python
rec = [100, 200]  # [w, h]

width = rec[0]
height = rec[1]

print(width, height)
```

```python
w, h = rec
```

```python
w, h
```

```python
w, h = (111, 222)

print(w, h)
```

```python
a, b, c = 10, 20, 30
```

```python
a, b, c = "XYZ"

print(a, b, c)
```
