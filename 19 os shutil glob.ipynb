# os 모듈: 디렉토리, 파일 관리

```python
import os
```

```python
savedir = "tmpdir1"
```

```python
if not os.path.exists(savedir):
  os.mkdir(savedir)
  print(savedir, '폴더 생성')

else:
  print(savedir, '이미 존재')
```

```python
savedir = r"tmpdir2/aaa/bbb/ccc"   # 경로는 raw-string 사용!
```

```python
if not os.path.exists(savedir):
  os.makedirs(savedir)
  print(savedir, '폴더 생성')

else:
  print(savedir, '이미 존재')
```

```python
os.listdir('./sample_data')
```

```python
os.getcwd()  # 현재 경로 (current working directory)
```

```python
# 경로 문자열 생성
os.path.join(os.getcwd(), 'aaa', 'bbb/ccc')
```
