# Argument Unpacking
í•¨ìˆ˜ **í˜¸ì¶œ** í• ë•Œ *

```python
def print_val(kor, eng, math):
    print(kor, eng, math)
    print('ì´ì =', kor + eng + math)

print_val(10, 20, 30)
```

```python
score = [11, 22, 33]
# print_val(score)

print_val(score[0], score[1], score[2])
```

```python
print_val(*score)  # argument unpacking ë°œìƒ!
```

```python
print_val(*(100, 200, 300))
```

```python
student = {"name": "Sam", "email": "sam@mail.com"}

def print_dict(name, email):
    print(name, email)
```

# filter() : (N' <= N)

```python
# êµ¬ë¬¸ : filter(function, iterable)

# filterì— ì¸ìë¡œ ì‚¬ìš©ë˜ëŠ” functionì€ ì²˜ë¦¬ë˜ëŠ” ê°ê°ì˜ ìš”ì†Œì— ëŒ€í•´ ì°¸/ê±°ì§“ ê°’ì„ ë°˜í™˜í•©ë‹ˆë‹¤.
# 'ì°¸'ì„ ë°˜í™˜í•˜ë©´ ê·¸ ìš”ì†ŒëŠ” ë‚¨ê²Œ ë˜ê³ , 'ê±°ì§“'ì„ ë°˜í™˜í•˜ë©´ ê·¸ ìš”ì†ŒëŠ” ì œê±° ë©ë‹ˆë‹¤

# filter() ê²°ê³¼ëŠ” filter ê°ì²´  (ì´ë˜í•œ iterable í•˜ë‹¤)
```

```python
def multiple3(x):
  return x % 3  == 0
```

```python
multiple3(8)
```

```python
multiple3(9)
```

```python
list(filter(multiple3, [3, 7, 9]))
```

```python
# ì…ë ¥: [3,    7,    9]
#       â†“      â†“    â†“
# ì ìš©:   multiple3()
#       â†“      â†“    â†“
#       T      F    T
#       â†“           â†“
# ê²°ê³¼: [3,         9]
```

```python
list(filter(lambda x: x % 3 == 0, [3, 7, 9]))
```

```python
dataset = [1, -2, 3, -4, 5]

# ì–‘ìˆ˜ì¸ ìˆ«ìë“¤ì„ ì œê³±í•œ ê²°ê³¼  => [1, 9, 25]
```

```python
list(filter(lambda x: x > 0, dataset))
```

```python
list(map(lambda x: x ** 2, filter(lambda x: x > 0, dataset)))
```

```python
list(map(
    lambda x: x ** 2,
    filter(lambda x: x > 0, dataset)))
```

```python
[x ** 2 for x in dataset if x > 0]
```

# global variable, global scope
í•¨ìˆ˜ ë°”ê¹¥ì—ì„œ ì„ ì–¸í•œ ë³€ìˆ˜.  

ìŠ¤í¬ë¦½íŠ¸ ì „ì²´ì—ì„œ ì ‘ê·¼í•  ìˆ˜ ìˆëŠ” ë³€ìˆ˜ë¥¼ ì „ì—­ ë³€ìˆ˜(global variable)ë¼ê³  ë¶€ë¦…ë‹ˆë‹¤.

íŠ¹íˆ ì „ì—­ ë³€ìˆ˜ì— ì ‘ê·¼í•  ìˆ˜ ìˆëŠ” ë²”ìœ„ë¥¼ ì „ì—­ ë²”ìœ„(global scope)ë¼ê³  í•©ë‹ˆë‹¤

```python
x = 10  # ì „ì—­ë³€ìˆ˜

def foo():
  print('foo() x =', x)  # x ëŠ” ì „ì—­ë³€ìˆ˜ x : <- 'ì½ê¸° ë™ì‘'

foo()

print(x)
```

```python
def goo():
  y = 10  # goo() ë‚´ì—ì„œ ì„ ì–¸í•œ ì§€ì—­ë³€ìˆ˜
  print('goo() y =', y)  # y ëŠ” goo() í•¨ìˆ˜ ì•ˆì—ì„œë§Œ ì‚¬ìš© ê°€ëŠ¥. ë¦¬í„´í•˜ë©´ ì†Œë©¸ë¨.

goo()
print(y)  # ë¦¬í„´í›„ì— y ê°’ ì—†ë‹¤!
```

```python
z = 10  # ì „ì—­ë³€ìˆ˜

def foo():
  z = 20  # z ëŠ” foo() ì˜ ì§€ì—­ë³€ìˆ˜!  'ì“°ê¸°ë™ì‘'
  print('foo() z =', z)

foo()

print(z)
```

```python
z = 10  # ì „ì—­ë³€ìˆ˜

def foo():
  global z  # z ëŠ” ì „ì—­ë³€ìˆ˜ z ë¥¼ ì˜ë¯¸
  z = 20
  print('foo() z =', z)

foo()

print(z)
```
