# в–  мһҗм—°м–ҙ мІҳлҰ¬ (NLP)
Natural Language Processing мқҙлһҖ

**мһҗм—°м–ҙ мІҳлҰ¬(NLP, Natural Language Processing)**лҠ” м»ҙн“Ён„°к°Җ мқёк°„мқҳ м–ём–ҙлҘј мқҙн•ҙн•ҳкі , н•ҙм„қн•ҳкі , мғқм„ұн•ҳлҠ” кё°мҲ мқ„ л§җн•©лӢҲлӢӨ. NLPлҠ” м–ём–ҙн•ҷкіј м»ҙн“Ён„° кіјн•ҷмқҳ мңөн•© л¶„м•јлЎң, м–ём–ҙлҘј мІҳлҰ¬н•ҳкі  л¶„м„қн•ҳлҠ” лӢӨм–‘н•ң мһ‘м—…мқ„ нҸ¬н•Ён•©лӢҲлӢӨ. к·ё лӘ©м ҒмқҖ мӮ¬лһҢкіј м»ҙн“Ён„° к°„мқҳ мғҒнҳёмһ‘мҡ©мқ„ мӣҗнҷңн•ҳкІҢ л§Ңл“Өм–ҙ мЈјлҠ” кІғмһ…лӢҲлӢӨ.

мһҗм—°м–ҙ мІҳлҰ¬мқҳ мЈјмҡ” мһ‘м—…мқҖ лӢӨмқҢкіј к°ҷмҠөлӢҲлӢӨ:

1. **нҳ•нғңмҶҢ л¶„м„қ (Morphological Analysis)**: л¬ёмһҘм—җм„ң лӢЁм–ҙлҘј лӮҳлҲ„кі , к°Ғ лӢЁм–ҙмқҳ нҳ•нғңлҘј л¶„м„қн•©лӢҲлӢӨ. мҳҲлҘј л“Өм–ҙ, н•ңкөӯм–ҙм—җм„ң "мӮ¬лһ‘н•ңлӢӨ"лқјлҠ” лӢЁм–ҙлҠ” "мӮ¬лһ‘"кіј "н•ңлӢӨ"лЎң лӮҳлҲ„м–ҙм§‘лӢҲлӢӨ.

1. **кө¬л¬ё л¶„м„қ (Syntactic Parsing)**: л¬ёмһҘмқҳ кө¬мЎ°лҘј л¶„м„қн•ҳкі , л¬ёлІ•м Ғ кҙҖкі„лҘј нҢҢм•…н•©лӢҲлӢӨ. мҳҲлҘј л“Өм–ҙ, "лӮҳлҠ” мӮ¬кіјлҘј лЁ№лҠ”лӢӨ"лқјлҠ” л¬ёмһҘм—җм„ң мЈјм–ҙ(лӮҳ), лӘ©м Ғм–ҙ(мӮ¬кіј), лҸҷмӮ¬(лЁ№лҠ”лӢӨ) кҙҖкі„лҘј мӢқлі„н•©лӢҲлӢӨ.

1. **мқҳлҜё л¶„м„қ (Semantic Analysis)**: л¬ёмһҘмқҙ м „лӢ¬н•ҳл ӨлҠ” мқҳлҜёлҘј м¶”лЎ н•©лӢҲлӢӨ. мҳҲлҘј л“Өм–ҙ, "л°°к°Җ кі н”„лӢӨ"лҠ” лӢЁмҲңнһҲ 'л°°'лқјлҠ” лӢЁм–ҙк°Җ м•„лӢҲлқј 'л°°'к°Җ 'н—Ҳкё°м§җ'мқ„ мқҳлҜён•ҳлҠ”м§ҖлҘј мқҙн•ҙн•©лӢҲлӢӨ.

1. **к°җм • л¶„м„қ (Sentiment Analysis)**: н…ҚмҠӨнҠём—җ нҸ¬н•Ёлҗң к°җм •мқ„ л¶„м„қн•ҳм—¬ кёҚм •м Ғ, л¶Җм •м Ғ, мӨ‘лҰҪм Ғ к°җм •мқ„ л¶„лҘҳн•©лӢҲлӢӨ. мҳҲлҘј л“Өм–ҙ, "мқҙ мҳҒнҷ”лҠ” м •л§җ мһ¬лҜёмһҲлӢӨ"лҠ” кёҚм •м Ғмқё к°җм •мқ„, "мқҙ мҳҒнҷ”лҠ” м§ҖлЈЁн•ҳлӢӨ"лҠ” л¶Җм •м Ғмқё к°җм •мқ„ лӮҳнғҖлғ…лӢҲлӢӨ.

1. **кё°кі„ лІҲм—ӯ (Machine Translation)**: н•ң м–ём–ҙм—җм„ң лӢӨлҘё м–ём–ҙлЎң лІҲм—ӯн•ҳлҠ” мһ‘м—…мһ…лӢҲлӢӨ. мҳҲлҘј л“Өм–ҙ, мҳҒм–ҙм—җм„ң н•ңкөӯм–ҙлЎң лІҲм—ӯн•ҳлҠ” мӢңмҠӨн…ңмқ„ к°ңл°ңн•ҳлҠ” кІғмһ…лӢҲлӢӨ.

1. **м§Ҳмқҳ мқ‘лӢө (Question Answering)**: мӮ¬мҡ©мһҗк°Җ мһ…л Ҙн•ң м§Ҳл¬ём—җ лҢҖн•ҙ м Ғм Ҳн•ң лӢөмқ„ м ңкіөн•ҳлҠ” мӢңмҠӨн…ңмһ…лӢҲлӢӨ. мҳҲлҘј л“Өм–ҙ, "нҢҢлҰ¬мқҳ мҲҳлҸ„лҠ” м–ҙл””мқёк°Җмҡ”?"лқјлҠ” м§Ҳл¬ём—җ лҢҖн•ҙ "нҢҢлҰ¬лҠ” н”„лһ‘мҠӨмқҳ мҲҳлҸ„мһ…лӢҲлӢӨ."лқјлҠ” лӢөмқ„ м ңкіөн•ҳлҠ” кІғмһ…лӢҲлӢӨ.

1. **н…ҚмҠӨнҠё мғқм„ұ (Text Generation)**: мЈјм–ҙм§„ мЈјм ңлӮҳ мЎ°кұҙм—җ л§һм¶° н…ҚмҠӨнҠёлҘј мһҗлҸҷмңјлЎң мғқм„ұн•ҳлҠ” мһ‘м—…мһ…лӢҲлӢӨ. мҳҲлҘј л“Өм–ҙ, мӮ¬мҡ©мһҗк°Җ м ңкіөн•ң мЈјм ңм—җ лҢҖн•ҙ мһҗлҸҷмңјлЎң кё°мӮ¬лҘј мһ‘м„ұн•ҳлҠ” кІғмқҙ мқҙм—җ н•ҙлӢ№н•©лӢҲлӢӨ.

мһҗм—°м–ҙ мІҳлҰ¬лҠ” лӢӨм–‘н•ң л¶„м•јм—җм„ң нҷңмҡ©лҗ©лӢҲлӢӨ. мҳҲлҘј л“Өм–ҙ, кІҖмғү м—”м§„, мқҢм„ұ мқёмӢқ мӢңмҠӨн…ң, мұ—лҙҮ, мһҗлҸҷ лІҲм—ӯ мӢңмҠӨн…ң л“ұм—җм„ң мӮ¬мҡ©лҗҳл©°, мөңк·јм—җлҠ” лҢҖк·ңлӘЁ м–ём–ҙ лӘЁлҚё(мҳҲ: GPT) л“ұмқҳ л°ңм „мңјлЎң мһҗм—°м–ҙ мІҳлҰ¬ кё°мҲ мқҙ лҚ”мҡұ м§„нҷ”н•ҳкі  мһҲмҠөлӢҲлӢӨ.

# в–  н…ҚмҠӨнҠё м „мІҳлҰ¬ (Text Preprocessing)

# 1. нҶ нҒ°нҷ” (Tokenization)

- мЈјм–ҙм§„ мҪ”нҚјмҠӨ(corpus)м—җм„ң нҶ нҒ°(token)мқҙлқј л¶ҲлҰ¬лҠ” лӢЁмң„лЎң лӮҳлҲ„лҠ” мһ‘м—…мқ„ нҶ нҒ°нҷ”(tokenization)лқјкі  н•©лӢҲлӢӨ

- нҶ нҒ°мқҳ лӢЁмң„лҠ” мғҒнҷ©м—җ л”°лқј лӢӨлҘҙлӮҳ, ліҙнҶө мқҳлҜёк°Җ мһҲлҠ” лӢЁмң„лЎң нҶ нҒ°мқ„ м •мқҳн•©лӢҲлӢӨ.

## лӢЁм–ҙ нҶ нҒ°нҷ”(Word Tokenization)
нҶ нҒ°мқҳ кё°мӨҖмқ„ лӢЁм–ҙ(word)лЎң н•ҳлҠ” кІҪмҡ°, лӢЁм–ҙ нҶ нҒ°нҷ”(word tokenization)лқјкі  н•©лӢҲлӢӨ. лӢӨл§Ң, м—¬кё°м„ң лӢЁм–ҙ(word)лҠ” лӢЁм–ҙ лӢЁмң„ мҷём—җлҸ„ лӢЁм–ҙкө¬, мқҳлҜёлҘј к°–лҠ” л¬ёмһҗм—ҙлЎңлҸ„ к°„мЈјлҗҳкё°лҸ„ н•©лӢҲлӢӨ.
мҳҲлҘј л“Өм–ҙліҙкІ мҠөлӢҲлӢӨ. м•„лһҳмқҳ мһ…л ҘмңјлЎңл¶Җн„° кө¬л‘җм җ(punctuation)кіј к°ҷмқҖ л¬ёмһҗлҠ” м ңмҷёмӢңнӮӨлҠ” к°„лӢЁн•ң лӢЁм–ҙ нҶ нҒ°нҷ” мһ‘м—…мқ„ н•ҙлҙ…мӢңлӢӨ. кө¬л‘җм җмқҙлһҖ л§Ҳм№Ён‘ң(.), м»ҙл§Ҳ(,), л¬јмқҢн‘ң(?), м„ёлҜёмҪңлЎ (;), лҠҗлӮҢн‘ң(!) л“ұкіј к°ҷмқҖ кё°нҳёлҘј л§җн•©лӢҲлӢӨ.


мһ…л Ҙ: **Time is an illusion. Lunchtime double so!**

````python
# KoNLPy  мҪ”м—”м—ҳнҢҢмқҙ http://konlpy.org/ko/latest/

# мҪ”м—”м—ҳнҢҢмқҙлҘј нҶөн•ҙм„ң мӮ¬мҡ©н•  мҲҳ мһҲлҠ” нҳ•нғңмҶҢ л¶„м„қкё°лЎң Okt(Open Korea Text), л©”мә…(Mecab),
# мҪ”лӘЁлһҖ(Komoran), н•ңлӮҳлҲ”(Hannanum), кј¬кј¬л§Ҳ(Kkma)к°Җ мһҲмҠөлӢҲлӢӨ.
````

````python
from konlpy.tag import Okt
from konlpy.tag import Kkma
````

````python
okt = Okt()
kkma = Kkma()
````

````python
text = "м—ҙмӢ¬нһҲ мҪ”л”©н•ң лӢ№мӢ , м—°нңҙм—җлҠ” м—¬н–үмқ„ к°Җлҙҗмҡ”"
````

````python
print("Okt")
print('нҳ•нғңмҶҢл¶„м„қ: ', okt.morphs(text))
print('н’ҲмӮ¬нғңк№…: ', okt.pos(text))
print('лӘ…мӮ¬м¶”м¶ң: ', okt.nouns(text))
````

````python
print("Kkma")
print('нҳ•нғңмҶҢл¶„м„қ: ', kkma.morphs(text))
print('н’ҲмӮ¬нғңк№…: ', kkma.pos(text))
print('лӘ…мӮ¬м¶”м¶ң: ', kkma.nouns(text))
````

# м •м ң(Cleaning), м •к·ңнҷ”(Normalization)

- **м •м ң(cleaning)** : к°–кі  мһҲлҠ” мҪ”нҚјмҠӨлЎңл¶Җн„° л…ёмқҙмҰҲ лҚ°мқҙн„°лҘј м ңкұ°н•ңлӢӨ.
- **м •к·ңнҷ”(normalization)** : н‘ңнҳ„ л°©лІ•мқҙ лӢӨлҘё лӢЁм–ҙл“Өмқ„ нҶөн•©мӢңмјңм„ң к°ҷмқҖ лӢЁм–ҙлЎң л§Ңл“Өм–ҙмӨҖлӢӨ.


````python
# (мқјл°ҳм ҒмңјлЎң)  м •м ң л°Ҹ м •к·ңнҷ”лҠ” 'нҶ нҒ°нҷ” мһ‘м—…м „ (м§Ғнӣ„)' м—җ мҲҳн–үн•ҳкіӨ н•ңлӢӨ
````

## н‘ңм ңм–ҙ м¶”м¶ң(Lemmatization)

````python
# н‘ңм ңм–ҙ(Lemma) : 'кё°ліё мӮ¬м „нҳ• лӢЁм–ҙ'
# н‘ңм ңм–ҙ м¶”м¶ңмқҖ лӢЁм–ҙл“Өмқҙ лӢӨлҘё нҳ•нғңлҘј к°Җм§ҖлҚ”лқјлҸ„, к·ё лҝҢлҰ¬ лӢЁм–ҙлҘј м°ҫм•„к°Җм„ң лӢЁм–ҙмқҳ к°ңмҲҳлҘј мӨ„мқј мҲҳ мһҲлҠ”м§Җ нҢҗлӢЁ
# ex) am, are, isлҠ” м„ңлЎң лӢӨлҘё мҠӨнҺ л§Ғмқҙм§Җл§Ң к·ё лҝҢлҰ¬ лӢЁм–ҙлҠ” be

# нҳ•нғңмҶҢ: 'мқҳлҜёлҘј к°Җм§„ к°ҖмһҘ мһ‘мқҖ лӢЁмң„'
# нҳ•нғңмҶҢмқҳ мў…лҘҳ: м–ҙк°„(stem)кіј м ‘мӮ¬(affix)

# 1) м–ҙк°„(stem)
# : лӢЁм–ҙмқҳ мқҳлҜёлҘј лӢҙкі  мһҲлҠ” лӢЁм–ҙмқҳ н•өмӢ¬ л¶Җл¶„.

# 2) м ‘мӮ¬(affix)
# : лӢЁм–ҙм—җ м¶”к°Җм Ғмқё мқҳлҜёлҘј мЈјлҠ” л¶Җл¶„.
````

## dictionary мӮ¬мҡ©н•ҳкё°

````python
raw_text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
raw_text
````

````python
# л¬ёмһҘ нҶ нҒ°нҷ”
sentences = sent_tokenize(raw_text)
sentences
````

````python
# мң„ кІ°кіјм—җ лӢЁм–ҙнҶ нҒ°нҷ” мҲҳн–ү
vocab = {}
preprocessed_sentences = []
stop_words = set(stopwords.words('english'))

for sentence in sentences:
  tokenized_sentence = word_tokenize(sentence)
  result = []

  for word in tokenized_sentence:
    word = word.lower()  # лӘЁл“  лӢЁм–ҙлҘј мҶҢл¬ёмһҗнҷ” н•ҳм—¬ лӢЁм–ҙмқҳ к°ңмҲҳ мӨ„мһ„.
    if word not in stop_words:  # нҶ нҒ°нҷ”лҗң лӢЁм–ҙ к°Җ л¶Ҳмҡ©м–ҙмқё кІҪмҡ° м ңкұ°
      if len(word) > 2:    # лӢЁм–ҙ кёёмқҙ 2 мқҙн•ҳмқё кІҪмҡ° м ңкұ°
        result.append(word)

        # л“ұмһҘл№ҲлҸ„мҲҳ мІҙнҒ¬
        if word not in vocab:
          vocab[word] = 0
        vocab[word] += 1

  preprocessed_sentences.append(result)

preprocessed_sentences
````

````python
# лӢЁм–ҙ м§‘н•© (мӮ¬м „)
vocab  # лӢЁм–ҙ: л№ҲлҸ„мҲҳ
````

````python
vocab['barber']
````

````python
# л№ҲлҸ„мҲҳк°Җ лҶ’мқҖ мҲңм„ңлҢҖлЎң м •л ¬
vocab_sorted = sorted(vocab.items(), key=lambda x:x[1], reverse=True)

vocab_sorted
````

````python
# м •мҲҳк°’ л¶Җм—¬  1 л¶Җн„° мӢңмһ‘.
word_to_index = {}

i = 0
for (word, frequency) in vocab_sorted:
  if frequency > 1:  # л№ҲлҸ„мҲҳк°Җ мһ‘мқҖ лӢЁм–ҙ м ңмҷё
    i += 1
    word_to_index[word] = i

word_to_index
````

````python
# л§Ңм•Ҫ л№ҲлҸ„мҲҳ мғҒмң„ 5к°ң лӢЁм–ҙл§Ң мӮ¬мҡ©н•ңлӢӨл©ҙ?

vocab_size = 5
words_frequency = [word for word, index in word_to_index.items() if index >= vocab_size + 1]

for w in words_frequency:
  del word_to_index[w]  # н•ҙлӢ№ лӢЁм–ҙм—җ лҢҖн•ң мқёлҚұмҠӨ м •ліҙлҘј мӮӯм ң

word_to_index
````

### OOV (Out-Of-Vocabulary)
мӮ¬м „м—җ м—ҶлҠ” (мқёлҚұмӢұ м•Ҳлҗң) лӢЁм–ҙнҶ нҒ°

````python
# OOV лҘј word_to_index  м—җ мғҲлЎӯкІҢ м¶”к°Җ
# лӢЁм–ҙм§‘н•©(мӮ¬м „) м—җ м—ҶлҠ” лӢЁм–ҙлҠ” 'OOV' лЎң мқёлҚұмӢұ н• кұ°лӢӨ.
````

````python
word_to_index['OOV'] = len(word_to_index) + 1

word_to_index
````

````python
# word_to_index лҘј мӮ¬мҡ©н•ҳм—¬ sentences мқҳ лӘЁл“  лӢЁм–ҙл“Өмқ„ м •мҲҳ мқёмҪ”л”© н•ҳмһҗ!
````

````python
preprocessed_sentences
````

````python
encoded_sentences = []

for sentence in preprocessed_sentences:
  encoded_sentence = []
  for word in sentence:
    try:
      encoded_sentence.append(word_to_index[word])
    except KeyError:
      encoded_sentence.append(word_to_index['OOV'])

  encoded_sentences.append(encoded_sentence)

encoded_sentences
````

````python
#  мӣҗліё л¬ёмһҘкіј мқёмҪ”л”© лҗң л¬ёмһҘмқ„ к°ҷмқҙ м¶ңл Ҙн•ҙліҙмһҗ

for e, s in zip(encoded_sentences, sentences):
  print(f'{str(e):25} {s}')

````

## мӣҗ-н•« мқёмҪ”л”© мһ‘м„ұ (мҲҳлҸҷ )

````python
# лӢЁм–ҙм§‘н•©мқҳ нҒ¬кё°лҘј лІЎн„°мқҳ 'м°Ёмӣҗ' мңјлЎң н•ҳкі 
# н‘ңнҳ„н•ҳкі  мӢ¶мқҖ лӢЁм–ҙмқҳ мқёлҚұмҠӨм—җ '1' к°’мқ„ л¶Җм—¬, лӢӨлҘё мқёлҚұмҠӨм—җлҠ” '0' мқ„ л¶Җм—¬н•ҳм—¬ лӢЁм–ҙмқҳ лІЎн„° н‘ңнҳ„
# мқҙл ҮкІҢ н‘ңнҳ„лҗң лІЎн„°лҘј мӣҗ-н•« лІЎн„° (one-hot vector)
````

````python
# нҶ нҒ°нҷ” мҲҳн–ү
tokens = okt.morphs('лӮҳлҠ” мһҗм—°м–ҙ мІҳлҰ¬лҘј л°°мҡҙлӢӨ')
tokens
````

````python
# кі мң н•ң м •мҲҳ л¶Җм—¬
word_to_index = {word: index for index, word in enumerate(tokens)}

word_to_index
````

````python
# нҶ нҒ°мқ„ мһ…л Ҙл°ӣм•„ мӣҗ-н•«лІЎн„° л§Ңл“ңлҠ” н•ЁмҲҳ
def one_hot_encoding(word, word_to_index):
  one_hot_vector = [0] * (len(word_to_index)) #лӢЁм–ҙм§‘н•©мқҳ нҒ¬кё°лҘј лІЎн„°мқҳ 'м°Ёмӣҗ' мңјлЎң н•ҳкі 
  index = word_to_index[word]  # н‘ңнҳ„н•ҳкі  мӢ¶мқҖ лӢЁм–ҙмқҳ мқёлҚұмҠӨм—җ '1' к°’мқ„ л¶Җм—¬
  one_hot_vector[index] = 1
  return one_hot_vector
````

````python
text = "лӮҳлһ‘ м җмӢ¬ лЁ№мңјлҹ¬ к°Ҳлһҳ м җмӢ¬ л©”лүҙлҠ” н–„лІ„кұ° к°Ҳлһҳ к°Ҳлһҳ н–„лІ„кұ° мөңкі м•ј"
# мң„мҷҖ к°ҷмқҖ л¬ёмһҘмқҙ мһҲлӢӨкі  н–Ҳмқ„ л•Ң, мјҖлқјмҠӨ нҶ нҒ¬лӮҳмқҙм ҖлҘј мқҙмҡ©н•ң м •мҲҳ мқёмҪ”л”©мқҖ лӢӨмқҢкіј к°ҷмҠөлӢҲлӢӨ.
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
print('лӢЁм–ҙ м§‘н•© :',tokenizer.word_index)
````

````python
sub_text = "м җмӢ¬ лЁ№мңјлҹ¬ к°Ҳлһҳ л©”лүҙлҠ” н–„лІ„кұ° мөңкі м•ј"
````

````python
encoded = tokenizer.texts_to_sequences([sub_text])[0]
encoded
````

````python
# мң„ м •мҲҳ мқёмҪ”л”© кІ°кіјлҘј мӣҗ-н•« мқёмҪ”л”© н•ҙліҙмһҗ
````

````python
one_hot = to_categorical(encoded)

one_hot
````
